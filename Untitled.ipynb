```python
db_active = pd.read_sql(db.session.query(Channel.uni_code, Channel.active).statement, con=db.session.bind)
```

```python
db_active = db_active.astype({'uni_code': 'string','active': 'int32'})
db_active.dtypes
```

```python
sl_active = sl_active.astype({'uni_code': 'string','active': 'int32'})
sl_active.dtypes
```

```python
db_active.loc[db_active.uni_code == 'AK_BMR:LHZ', 'active'] = 0
db_active = db_active[:-2]
```

```python
active = sl_active.join(db_active, how='outer', lsuffix='_sl')
```

```python
df = sl_active.join(db_active, how='outer', lsuffix='_sl')
df.loc[df.uni_code.isnull(), 'uni_code'] = \
    df.loc[df.uni_code.isnull(), 'uni_code_sl']
df['active_db'] = df.active.fillna(0)
df.active_sl.fillna(0, inplace=True)
df
```

```python
df['diff'] = df.active_sl - df.active_db
df = df[df['diff'] != 0]
df.loc[df['diff'] == -1, 'diff'] = 0
# format for database
df = df[['uni_code','diff']].reset_index(drop=True)
df['time'] = sl.access_time
df
```
